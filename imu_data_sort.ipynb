```python
import numpy as np
import pandas as pd
import os
```

```python
df_is_imu = pd.read_csv("data/is_imu_data.csv")
df_is_imu.head()
```

```python
df_lidar_imu = pd.read_csv("data/lidar_imu_data.csv")
df_lidar_imu.head()
```

```python
df_sorted = pd.concat([df_is_imu, df_lidar_imu[4:]])
df_sorted = df_sorted.sort_values(by="time")
df_sorted
```

```python
np_sorted = df_sorted.to_numpy()
```

```python
for s in np_sorted:
    print(s[0])
```

```python
merged = pd.merge_asof(
    df_lidar_imu[4:],
    df_is_imu,
    on="time",
    direction="nearest"
)
merged.head
```

```python
print(np.shape(df_is_imu))
print(np.shape(df_lidar_imu))
print(np.shape(merged))
```

```python
def save_df_to_csv(df: pd.DataFrame, save_name: str):
    file_path = "data"
    if not os.path.exists(file_path):
        os.makedirs(file_path)
    file_path = os.path.join(file_path, save_name+".csv")
    df.to_csv(file_path, index=False)

save_df_to_csv(merged, "sorted_imu_data")
```
